This notebook analyzes the famous **Fisher's Iris dataset**, which contains measurements of **sepal length, sepal width, petal length, and petal width** for three species of iris flowers: **Setosa, Versicolor, and Virginica**.


```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv")

# Save dataset locally
df.to_csv("iris.csv", index=False)

# Display the dataset in Jupyter Notebook
pd.set_option('display.max_rows', None)  # Show all rows
pd.set_option('display.max_columns', None)  # Show all columns
df
```

### Step 4: Generate and Save Variable Summaries


```python
# Create summary for numerical variables
numerical_summary = df.describe(include=[np.number])

# Create summary for categorical variables
categorical_summary = df.describe(include=[object])

# Save summaries to a file
with open('Summary_variable.txt', 'w') as file:
    file.write("Numerical Variables Summary:\n")
    file.write(numerical_summary.to_string())
    file.write("\n\nCategorical Variables Summary:\n")
    file.write(categorical_summary.to_string())

# Display summaries in notebook
numerical_summary, categorical_summary
```

### Step 5: Create Histograms for Each Variable
These histograms show the distribution of each variable in the dataset.


```python
# Define histogram function
def plot_histogram(column, color):
    plt.hist(df[column], bins=5, color=color, edgecolor='black')
    plt.xlabel(column)
    plt.title(f'Histogram of {column.capitalize()}')
    plt.show()

# Plot histograms
plot_histogram('species', 'thistle')
plot_histogram('sepal_length', 'rebeccapurple')
plot_histogram('sepal_width', 'rebeccapurple')
plot_histogram('petal_length', 'rebeccapurple')
plot_histogram('petal_width', 'rebeccapurple')
```

### Step 6: Scatter Plots of Variable Pairs by Species
 These scatter plots visualize relationships between different numerical variables, categorized by species.


```python
# Function to create scatter plots
def plot_scatter(x_col, y_col):
    plt.figure(figsize=(10, 8))
    sns.scatterplot(data=df, x=x_col, y=y_col, hue='species', palette=['indigo', 'green', 'yellow'])
    plt.title(f'Scatter Plot of {x_col.capitalize()} vs {y_col.capitalize()} by Species')
    plt.show()

# Scatter plots for variable pairs
plot_scatter('sepal_length', 'sepal_width')
plot_scatter('petal_length', 'petal_width')
plot_scatter('sepal_length', 'petal_length')
plot_scatter('sepal_length', 'petal_width')
plot_scatter('sepal_width', 'petal_length')
plot_scatter('sepal_width', 'petal_width')
```

# ## Step 7: Correlation Analysis & Heatmap
# The correlation matrix shows the relationships between numerical variables.


```python
# Compute correlation matrix
correlation_matrix = df.select_dtypes(include=['float64', 'int64']).corr().round(2)

# Save correlation matrix
with open('correlation_matrix.txt', 'w') as file:
    file.write("Correlation Matrix:\n")
    file.write(correlation_matrix.to_string())

# Display correlation matrix in Jupyter Notebook
correlation_matrix
```

```python
# Create heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix Heatmap")
plt.show()
```
